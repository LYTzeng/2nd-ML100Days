 假設我們資料中類別的數量並不均衡，在評估準確率時可能會有所偏頗，試著切分出 y_test 中，0 類別與 1 類別的數量是一樣的 (亦即 y_test 的類別是均衡的)

```python
import numpy as np
X = np.arange(1000).reshape(200, 5)
y = np.zeros(200)
y[:40] = 1

```

```python
y
```

 可以看見 y 類別中，有 160 個 類別 0，40 個 類別 1 ，請試著使用 train_test_split 函數，切分出 y_test 中能各有 10 筆類別 0 與 10 筆類別 1 。(HINT: 參考函數中的 test_size，可針對不同類別各自作切分後再合併)

```python
def split(X, y):
    y_zero_idx = np.where(y == 0)
    X_class0, y_class0 = X[y_zero_idx], y[y_zero_idx]

    y_one_idx = np.where(y == 1)
    X_class1, y_class1 = X[y_one_idx], y[y_one_idx]

    return X_class0, y_class0, X_class1, y_class1
```

```python
from sklearn.model_selection import train_test_split
X_class0, y_class0, X_class1, y_class1 = split(X, y)

X_train0, X_test0, y_train0, y_test0 = train_test_split(X_class0, y_class0, test_size=1/16, random_state=42)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_class1, y_class1, test_size=1/4, random_state=42)
X_train = np.concatenate((X_train0, X_train1))
X_test = np.concatenate((X_test0, X_test1))
y_train = np.concatenate((y_train0, y_train1))
y_test = np.concatenate((y_test0, y_test1))

print(y_test)
```
